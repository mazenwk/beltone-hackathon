```python
import pandas as pd 
import matplotlib.pyplot as plt
import itertools
```

```python
crude_oil_prices = pd.read_csv('../InputData/crude_oil_prices.csv')
federal_rates = pd.read_csv('../InputData/effective_federal_funds_rate.csv')
corridor_rates = pd.read_csv('../InputData/egyptian_corridor_interest_rates.csv')
housing_index = pd.read_csv('../InputData/housing_index.csv')
inflation_mom = pd.read_csv('../InputData/inflation_month_on_month.csv')
inflation_yoy = pd.read_csv('../InputData/inflation_year_on_year.csv')
news_data = pd.read_csv('../InputData/news.csv')
stock_prices = pd.read_csv('../InputData/stocks_prices_and_volumes.csv')
vix_indices = pd.read_csv('../InputData/vix_index.csv')
vixeem_indices = pd.read_csv('../InputData/vxeem_index.csv')
gold_prices = pd.read_csv('../InputData/intraday_gold.csv')
gold_prices_target = pd.read_csv('../InputData/target_gold.csv')
```

## gold prices target

```python
gold_prices_target.head()
```

```python
import matplotlib.dates as mdates

# Ensure the 'Date' column is in datetime format
gold_prices_target['date'] = pd.to_datetime(gold_prices_target['date'])

# Calculate FOB_EBCO
# FOB_EBCO = crude_oil_prices.iloc[:, 1] / crude_oil_prices.iloc[:, 2]

# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(gold_prices_target['date'], gold_prices_target['pct_change'] , color='blue', alpha=0.6)

# Set the title and labels
plt.title('pct_change Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('FOB_EBCO', fontsize=14)

# Format the x-axis to show only the years
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Set major ticks to every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format ticks as 'YYYY'

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid and adjust layout
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

## Crude Oil Prices


```python
crude_oil_prices.head()
```

```python
import matplotlib.dates as mdates

# Ensure the 'Date' column is in datetime format
crude_oil_prices['Date'] = pd.to_datetime(crude_oil_prices['Date'])

# Calculate FOB_EBCO
FOB_EBCO = crude_oil_prices.iloc[:, 1] / crude_oil_prices.iloc[:, 2]

# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(crude_oil_prices['Date'], FOB_EBCO, color='blue', alpha=0.6)

# Set the title and labels
plt.title('FOB_EBCO Prices Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('FOB_EBCO', fontsize=14)

# Format the x-axis to show only the years
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Set major ticks to every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format ticks as 'YYYY'

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid and adjust layout
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Get the list of column names
columns = crude_oil_prices.columns

# Plot each column against every other column
for col1, col2 in itertools.combinations(columns, 2):
    print(col1)
    plt.figure()
    plt.scatter(crude_oil_prices[col1], crude_oil_prices[col2])
    plt.title(f'{col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
    []
```
